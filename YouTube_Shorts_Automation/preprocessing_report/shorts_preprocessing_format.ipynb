```python
import pandas as pd
import numpy as np
# Assuming df is your dataframe containing the data
df = pd.read_excel("Buzz Wave/ì™¸ì£¼ ê²°ê³¼/phase123.xlsx")
```

```python
# Released Date ì „ì²˜ë¦¬: Convert string dates to datetime format with custom format
df['Released Date'] = pd.to_datetime(df['Released Date'], format='%b %d %Y')
# Convert datetime format to yyyy-mm-dd string format
df['Released Date'] = df['Released Date'].dt.strftime('%Y-%m-%d')
```

```python

#  Check Date ì „ì²˜ë¦¬: column to datetime
df['Check Date'] = pd.to_datetime(df['Check Date'], errors='coerce')

# Check if values are already in the desired format
already_formatted = df['Check Date'].dt.strftime('%Y-%m-%d') == df['Check Date']

# Convert only the values not already in the desired format
df.loc[~already_formatted, 'Check Date'] = pd.to_datetime(df.loc[~already_formatted, 'Check Date'], errors='coerce')

# Convert datetime objects to the desired format "yyyy-mm-dd"
df['Check Date'] = df['Check Date'].dt.strftime('%Y-%m-%d')
```

```python
# subscribers ì „ì²˜ë¦¬: Function to convert subscriber counts to numerical values
def convert_subscriber_count(subscriber_str):
    if 'K' in subscriber_str:
        subscriber_str = subscriber_str.replace('K', '')
        subscriber_str = subscriber_str.replace(' subscribers', '')
        return int(float(subscriber_str) * 1000)
    elif 'M' in subscriber_str:
        subscriber_str = subscriber_str.replace('M', '')
        subscriber_str = subscriber_str.replace(' subscribers', '')
        return int(float(subscriber_str) * 1000000)
    else:
        # Split the string to separate the number and unit, then convert to float
        num, _ = subscriber_str.split()
        return int(float(num))

# Apply the function to the Subscribers column
df['subscribers'] = df['subscribers'].apply(convert_subscriber_count)
```

```python
# Likes ì „ì²˜ë¦¬
def convert_likes_count(likes_str):
    
    try:
        if 'K' in likes_str:
            likes_str = likes_str.replace('K\n', '')
            return int(float(likes_str) * 1000)
        elif 'M' in likes_str:
            likes_str = likes_str.replace('M\n', '')
            return int(float(likes_str) * 1000000)
        else:
             return int(float(likes_str))
    except Exception:
        pass
df['Likes'] = df['Likes'].apply(convert_likes_count)
df['Likes'] = df['Likes'].fillna(0)
df['Likes'] = df['Likes'].replace('None', 0)
```

```python
# Views ì „ì²˜ë¦¬
df['Views'] = df['Views'].str.replace('\n', '')
df['Views'] = df['Views'].str.replace(',', '')
df['Views'] = df['Views'].replace('NaN', 0)
df['Views'] = df['Views'].replace('None', np.nan)
df['Views'] = df['Views'].fillna(0)
df['Views'] = df['Views'].astype(int)
df['Views'].unique()
```

```python
df.sort_values(by=['Shorts Name', 'link', 'Check Date'], inplace = True)
```

```python
df['Views Diff'] = df.groupby('Shorts Name')['Views'].diff().fillna(0)
df['Likess Diff'] = df.groupby('Shorts Name')['Likes'].diff().fillna(0)
df
```
