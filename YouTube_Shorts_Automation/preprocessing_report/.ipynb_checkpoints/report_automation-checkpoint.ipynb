```python
# 되기는 하는거
from reportlab.lib import colors as rl_colors
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import inch
from reportlab.graphics.shapes import Drawing, Rect, String
import random

# Sample data
data = [
    ("USA", 331449281),
    ("China", 1444216107),
    ("India", 1393409038),
    ("Brazil", 212559417),
    ("Pakistan", 225199937),
]

# Colors for the stack graph bars
colors = [rl_colors.blue, rl_colors.green, rl_colors.red, rl_colors.orange, rl_colors.purple]

# Function to create a stack graph
def create_stack_graph():
    drawing = Drawing(400, 200)
    data = [[random.randint(100, 500) for _ in range(5)] for _ in range(3)]
    categories = [str(i) for i in range(1, 6)]
    max_value = max([sum(row) for row in data])
    num_categories = len(data[0])

    # Draw bars
    bar_width = 50
    bar_gap = 20
    bar_x = 50
    for i in range(num_categories):
        bar_height = 0
        for row in data:
            bar_height += row[i] / max_value * 125
            drawing.add(Rect(bar_x, 50, bar_width, bar_height, fillColor=colors[i % len(colors)], strokeColor=colors[i % len(colors)]))
        bar_x += bar_width + bar_gap

    # Draw legend
    legend_x = 350
    legend_y = 175
    legend_spacing = 20
    for i, color in enumerate(colors):
        drawing.add(Rect(legend_x, legend_y - i * legend_spacing, 10, 10, fillColor=color, strokeColor=color))
        drawing.add(String(legend_x + 15, legend_y - i * legend_spacing - 5, categories[i], fontSize=10, fillColor=colors[i % len(colors)]))

    return drawing

# Create a table of country data
def create_country_table(data):
    table_data = [["Country", "Population"]]
    for country, population in data:
        table_data.append([country, str(population)])
    table_style = TableStyle([('BACKGROUND', (0, 0), (-1, 0), rl_colors.lightgrey),
                              ('TEXTCOLOR', (0, 0), (-1, 0), rl_colors.black),
                              ('ALIGN', (0, 0), (-1, -1), 'CENTER'),
                              ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
                              ('BOTTOMPADDING', (0, 0), (-1, 0), 12),
                              ('BACKGROUND', (0, 1), (-1, -1), rl_colors.beige),
                              ('GRID', (0, 0), (-1, -1), 1, rl_colors.black)])
    t = Table(table_data)
    t.setStyle(table_style)
    return t

# Create a description paragraph
def create_description():
    description = """
    This report contains information about the population of various countries. 
    It includes a stack graph showing distribution and a table listing the countries 
    and their respective populations.
    """
    return Paragraph(description, getSampleStyleSheet()["BodyText"])

# Create the PDF
def create_report():
    doc = SimpleDocTemplate("report.pdf", pagesize=letter)
    Story = []
    styles = getSampleStyleSheet()
    styles.add(ParagraphStyle(name='Justify', alignment=1))
    
    # Add margin
    Story.append(Spacer(1, 2*inch))

    # Add stack graph
    Story.append(create_stack_graph())

    # Add margin
    Story.append(Spacer(1, 0.5*inch))

    # Add table
    Story.append(create_country_table(data))

    # Add description
    Story.append(Spacer(1, 0.5*inch))
    Story.append(create_description())

    doc.build(Story)

if __name__ == "__main__":
    create_report()
```

### weekly report V0.1

```python
import random
import pandas as pd
from reportlab.lib.pagesizes import A4
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib import colors as rl_colors
from reportlab.graphics.shapes import Drawing, String, Line
from reportlab.lib.units import inch

# Existing data
data = {
    "short name": ["Short 1", "Short 2", "Short 3", "Short 4", "Short 5"],
    "1week past": [100, 80, 120, 90, 110],
    "present": [150, 120, 180, 130, 160],
    "channel url": ["channel_url_1", "channel_url_2", "channel_url_3", "channel_url_4", "channel_url_5"],
    "channel name": ["Channel 1", "Channel 2", "Channel 3", "Channel 4", "Channel 5"],
    "category": ["people-blogs", "people-blogs", "people-blogs", "people-blogs", "people-blogs"],
    "short url": ["short_url_1", "short_url_2", "short_url_3", "short_url_4", "short_url_5"]
}

# Additional data for "Music" category
music_data = {
    "short name": [f"Music Short {i}" for i in range(1, 6)],
    "1week past": [110, 90, 130, 100, 120],
    "present": [160, 130, 170, 140, 150],
    "channel url": [f"music_channel_url_{i}" for i in range(1, 6)],
    "channel name": [f"Music Channel {i}" for i in range(1, 6)],
    "category": ["music"] * 5,
    "short url": [f"music_short_url_{i}" for i in range(1, 6)]
}

# Additional data for "how-to-style" category
how_to_style_data = {
    "short name": [f"How-To Short {i}" for i in range(1, 6)],
    "1week past": [90, 70, 100, 80, 110],
    "present": [140, 120, 150, 130, 160],
    "channel url": [f"how_to_style_channel_url_{i}" for i in range(1, 6)],
    "channel name": [f"How-To Style Channel {i}" for i in range(1, 6)],
    "category": ["how-to-style"] * 5,
    "short url": [f"how_to_style_short_url_{i}" for i in range(1, 6)]
}

# Extend existing data to reach 15 examples
for i in range(6, 16):
    data["short name"].append(f"Short {i}")
    data["1week past"].append(random.randint(80, 120))
    data["present"].append(random.randint(130, 170))
    data["channel url"].append(f"channel_url_{i}")
    data["channel name"].append(f"Channel {i}")
    data["category"].append("people-blogs")
    data["short url"].append(f"short_url_{i}")

# Combine all the data
df = pd.DataFrame(data)

# Concatenate additional data for "Music" and "how-to-style" categories
df = pd.concat([df, pd.DataFrame(music_data), pd.DataFrame(how_to_style_data)], ignore_index=True)

# Function to format large numbers as K, M, or B
def format_number(value):
    magnitude = 0
    while abs(value) >= 1000:
        magnitude += 1
        value /= 1000.0
    # Format the number with appropriate suffix
    if magnitude == 1:
        return f'{value:.1f}K'
    elif magnitude == 2:
        return f'{value:.1f}M'
    elif magnitude == 3:
        return f'{value:.1f}B'
    else:
        return str(value)

# Function to create a line graph for a category
def create_line_graph(category_data, short_name):
    drawing = Drawing(200, 120)  # Adjusted size of the drawing
    x_values = [40, 80, 120, 160, 200]  # Points on x-axis (week0, week1, week2, week3, 1Month)
    
    # Normalize the data to the same scale
    max_y_value = max(category_data["week0"], category_data["week1"], category_data["week2"], category_data["week3"], category_data["1Month"])
    
    y_values = [
        category_data["week0"] / max_y_value,
        category_data["week1"] / max_y_value,
        category_data["week2"] / max_y_value,
        category_data["week3"] / max_y_value,
        category_data["1Month"] / max_y_value
    ]
    
    drawing.add(Line(40, 16, 200, 16))  # Adjusted x-axis length
    drawing.add(Line(40, 16, 40, 96))  # Adjusted y-axis length
    
    for i in range(4):  # Draw lines between week0 and Month1
        x1 = x_values[i]
        y1 = 16 + (y_values[i] * 80)
        x2 = x_values[i + 1]
        y2 = 16 + (y_values[i + 1] * 80)
        drawing.add(Line(x1, y1, x2, y2, strokeWidth=0.5, strokeColor=rl_colors.blue))
    
    drawing.add(String(20, 110, short_name, fontSize=8, textAnchor="middle"))
    
    # Format the y-axis labels
    for i in range(6):
        value = max_y_value * i / 5
        formatted_value = format_number(value)
        drawing.add(String(10, 16 + 80 * i - 5, formatted_value, fontSize=8))
    
    return drawing

# Function to create the PDF report
def create_report():
    doc = SimpleDocTemplate("YouTube_Shorts_Report.pdf", pagesize=A4)
    Story = []
    styles = getSampleStyleSheet()
    for category, category_data in df.groupby('category'):
        category_title = Paragraph(category.capitalize(), styles["Title"])
        Story.append(category_title)
        Story.append(Spacer(1, 0.2 * inch))
        for _, short_data in category_data.head(5).iterrows():
            short_graph = create_line_graph(short_data, short_data["short name"])
            view_increase = short_data["present"] - short_data["1week past"]
            description = f"Channel name: {short_data['channel name']}\nView increase over 7 days: {view_increase}\nShort URL: {short_data['short url']}"
            category_table_data = [[short_graph, description]]
            category_table = Table(category_table_data)
            category_table.setStyle(TableStyle([('INNERGRID', (0, 0), (-1, -1), 0.25, rl_colors.black),
                                                 ('BOX', (0, 0), (-1, -1), 0.25, rl_colors.black)]))
            Story.append(category_table)
    doc.build(Story)

# Generate the report and save it as PDF
create_report()
```
