```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import undetected_chromedriver as uc
import numpy as np
import random  # Random delay
import re  # Regular expressions for cleaning
from datetime import datetime, timedelta

today = datetime.today().strftime('%Y-%m-%d')
# Function to scroll to the bottom of the page
def scroll_to_bottom(driver):
    last_height = driver.execute_script("return document.documentElement.scrollHeight")
    while True:
        driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)  # Scroll down
        time.sleep(random.uniform(2, 3))  # Allow time for new content to load
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Function to log into Playboard
def login(driver):
    driver.get("https://playboard.co/")
    time.sleep(random.uniform(2, 3))  # Wait for the page to load

    # Click the login button
    login_button = driver.find_element(By.XPATH, "//*[@id='app']/div[1]/header/div[5]/div[2]/a")
    login_button.click()
    time.sleep(random.uniform(2, 3))  # Wait for the login form to appear

    # Enter email
    email_input = driver.find_element(By.XPATH, "//*[@id='app']/div[1]/div/main/div[2]/form/input[1]")
    email_input.send_keys("dnova@d-nova.com")
    
    # Enter password
    password_input = driver.find_element(By.XPATH, "//*[@id='app']/div[1]/div/main/div[2]/form/input[2]")
    password_input.send_keys("hjl960214!!")
    
    # Click the submit button
    submit_button = driver.find_element(By.XPATH, "//*[@id='app']/div[1]/div/main/div[2]/form/button")
    submit_button.click()
    time.sleep(random.uniform(2, 3))  # Wait for login to complete

# Function to clean data before saving to Excel
def clean_data(value):
    if isinstance(value, str):
        # Remove illegal characters
        return re.sub(r'[<>:"/\\|?*]', '', value)
    return value

# Function to collect data and create a DataFrame
def playboard_collection(url, driver):
    try:
        driver.get(url)
        driver.implicitly_wait(10)  # Wait for the initial content to load
        
        # Scroll down to load all data
        scroll_to_bottom(driver)

        # BeautifulSoup로 HTML 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # 테이블 선택
        table = soup.select_one('table')  # 테이블 선택

        if table is None:
            raise ValueError("테이블을 찾을 수 없습니다. 페이지 구조를 확인하세요.")

        # 데이터 수집
        link_data = []
        for row in table.select('tr')[1:]:  # 첫 번째 행은 헤더이므로 제외
            link_element = row.select_one('td:nth-child(5) a')  # 5번째 열의 a 태그 찾기
            
            # href와 title 값 추출
            if link_element:
                link_href = link_element['href']
                link_title = link_element['title'] if 'title' in link_element.attrs else ''
                link_data.append((link_title, link_href))

        return link_data  # title과 href 반환

    except Exception as e:
        print(f"오류 발생: {e}")
        return []

# Function to create DataFrame with required columns
def create_dataframe(url, link_data):
    # Initialize category and country as NaN
    category = np.nan
    country = np.nan

    # Attempt to extract category and country from the URL
    try:
        # Extract country part
        if "south-korea" in url:
            country = "South Korea"
        elif "united-states" in url:
            country = "United States"

        # Use the URL to determine the category directly
        category_part = url  # Use the full URL as category_part

        # Determine the category based on the category_part
        if "all-videos" in category_part:
            category = "all"
        elif "animals" in category_part:
            category = "pets-animals"
        elif "music" in category_part:
            category = "music"
        elif "gaming" in category_part:
            category = "gaming"
        elif "vlog" in category_part:
            category = "people-blogs"
        elif "travel" in category_part:
            category = "travel-events"
        elif "sports" in category_part:
            category = "sports"
        elif "vehicles" in category_part:
            category = "autos-vehicles"
        elif "comedy" in category_part:
            category = "comedy"
        elif "entertainment" in category_part:
            category = "entertainment"
        elif "film" in category_part:
            category = "film-animation"
        elif "howto" in category_part:
            category = "howto-style"
        elif "education" in category_part:
            category = "education"
        elif "science" in category_part:
            category = "science-technology"
        elif "news" in category_part:
            category = "news-politics"
        elif "shows" in category_part:
            category = "shows"

    except Exception as e:
        print(f"URL 형식이 예상과 다릅니다: {e}")

    # Prepare the data for DataFrame
    data = []
    for channel_name, yt_fixed_url in link_data:
        # Extract channel ID from the href
        channel_id = yt_fixed_url.split('/')[-1]  # Assuming yt_fixed_url contains the channel ID
        yt_url = f"https://www.youtube.com/channel/{channel_id}/shorts"  # Construct the full URL

        data.append([
            np.nan,           # youtuberme_url
            category,        # category
            country,         # country
            clean_data(channel_name),  # Cleaned channel_name
            np.nan,          # subscribers
            np.nan,          # total_video_views
            np.nan,          # total_video_count
            np.nan,          # started
            np.nan,          # yt_url
            yt_url,          # yt_fixed_url
            'y'              # data_yn
        ])

    # Create DataFrame
    df = pd.DataFrame(data, columns=[
        'youtuberme_url', 'category', 'country', 'channel_name',
        'subscribers', 'total_video_views', 'total_video_count',
        'started', 'yt_url', 'yt_fixed_url', 'data_yn'
    ])
    
    return df


# Shuffle the shorts_url_list to randomize the order
random.shuffle(shorts_url_list)
shorts_url_list = shorts_url_list
# Initialize an empty DataFrame to hold all results
combined_df = pd.DataFrame()

driver = uc.Chrome()  # Initialize the Selenium WebDriver with headless option options=uc.ChromeOptions().add_argument("--headless")

# Attempt to log in and process each URL
for url in shorts_url_list:
    try:
        # login(driver)  # Log in to the site
        link_info = playboard_collection(url, driver)
        
        # Create DataFrame from the collected data
        df = create_dataframe(url, link_info)
        
        # Append the new DataFrame to the combined DataFrame
        combined_df = pd.concat([combined_df, df], ignore_index=True)  # Append the new DataFrame

    except Exception as e:
        print(f"오류 발생: {e}. 다시 로그인 시도합니다...")
        combined_df.to_excel('combined_results.xlsx', index=False)  # Save the DataFrame to an Excel file
        driver.quit()  # Quit the driver
        time.sleep(random.uniform(120, 150))  # Wait before retrying
        
        driver = uc.Chrome()  # Reinitialize the WebDriver options=uc.ChromeOptions().add_argument("--headless")
        # login(driver)  # Log in to the site again
        
        link_info = playboard_collection(url, driver)
        
        # Create DataFrame from the collected data
        df = create_dataframe(url, link_info)
        
        # Append the new DataFrame to the combined DataFrame
        combined_df = pd.concat([combined_df, df], ignore_index=True)  # Append the new DataFrame

driver.quit()
# Save the final combined DataFrame to Excel
combined_df.to_excel('combined_results.xlsx', index=False)

# 결과 출력
print(combined_df)
```
